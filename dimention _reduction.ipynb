## univariate analysis by using hist and box plot

### To import necessary packages and code for plots

```python
import matplotlib.pyplot as plt
import seaborn as sb
for column in columnames:
    print(column)
    plt.figure(figsize=(17,6))
    plt.subplot(1,2,1)
    data[column].hist(bins=15,color='green')
    #plt.ylable("count")
    plt.subplot(1,2,2)
    sb.boxplot(x=data[column])
    plt.show()
```
